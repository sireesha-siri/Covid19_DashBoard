```python
import dash
import dash_html_components as html 
import dash_core_components as dcc
import plotly.graph_objects as go
from dash.dependencies import Input, Output

app = dash.Dash()

colors = {'background' : 'violet', 'text':'#7FDBFF'}

days=['18/1','19/1','20/1','21/1','22/1','23/1','24/1','25/1',
      '26/1','27/1','28/1','29/1','30/1','31/1','1/2','2/2',
      '3/2','4/2','5/2','6/2','7/2','8/2','9/2','10/2','11/2',
     '12/2','13/2','14/2','15/2','16/2','17/2','18/2','19/2',
     '20/2','21/2','22/2','23/2','24/2','25/2','26/2','27/2',
     '28/2','1/3','2/3','3/3','4/3','5/3','6/3','7/3','8/3',
     '9/3','10/3','11/3','12/3','13/3','14/3','15/3','16/3',
      '17/3','18/3']
confcases=[134,122,157,143,154,130,150,130,99,138,113,70,
               108,114,79,60,71,87,97,82,72,72,105,92,95,58,59,
               73,88,60,75,63,53,65,58,69,62,68,89,99,86,87,86,
                70,97,81,10,54,70,59,57,69,63,64,65,71,66,65,
                82,84]
fig1=go.Figure()

category=['Active','Recovered','Deceased']
ccases=[595,335923,1918]

age=['0-14','15-40','41-60','Above 60']
activecases=[17166,173372,110692,37259]

gender=['Male','Female']
gcases=[223923,115030]
fig2=go.Figure()


day=['18/1','19/1','20/1','21/1','22/1','23/1','24/1','25/1','26/1','27/1',
     '28/1','29/1','30/1','31/1','1/2','2/2','3/2','4/2','5/2','6/2',
     '7/2','8/2','9/2','10/2','11/2','12/2','13/2','14/2','15/2',
     '16/2','17/2','18/2','19/2','20/2','21/2','22/2','23/2','24/2',
    '25/2','26/2','27/2','28/2','1/3','2/3','3/3','4/3','5/3','6/3',
     '7/3','8/3','9/3','10/3','11/3','12/3','13/3','14/3','15/3',
     '16/3','17/3','18/3']
cases=[25,26,30,17,18,14,22,32,20,15,17,9,12,21,10,8,11,26,16,
       15,14,11,14,11,20,10,9,5,27,9,11,10,7,14,8,11,5,6,8,8,10,6,6,13,
       11,6,13,7,12,6,7,14,7,10,13,4,7,15,7,10] 
fig3=go.Figure()  

cat=['Active','Recovered','Deceased']
case=[87,15655,173] 
fig4=go.Figure()      
 

app.layout=html.Div([html.Div(html.H1('Covid 19 Dashboard',
                                      style={'color':'white',
                                             'text-align':'center',
                                             'background-color':'gray',
                                             'border-style':'double',
                                             'display':'inline-block',
                                             'width':'100%'})),
                     html.Div(dcc.Dropdown(id='dropdown1',
                                           options=[{'label':'Odisha Status',
                                                     'value':'os'},
                                                    {'label':'Bolangir District Status',
                                                     'value':'bds'}],
                                           placeholder="Select an option")),
                     html.Div([dcc.Graph(id='graph1',figure=fig1,
                                         style={'width':'49%','display':'inline-block'}
                                         ),
                               dcc.Graph(id='graph2',figure=fig2,
                                         style={'width':'49%','display':'inline-block'})]),
                     html.Div([dcc.Graph(id='graph3',figure=fig3,
                                         style={'width':'49%','display':'inline-block'}),
                               dcc.Graph(id='graph4',figure=fig4,
                                         style={'width':'49%','display':'inline-block'})])])

@app.callback(
             Output('graph1','figure'),Output('graph2','figure'),
               Output('graph3','figure'),Output('graph4','figure'),
              [Input('dropdown1','value')])

def update_figure(dropdown1):
    global fig1,fig2,fig3,fig4
    
    if (dropdown1=='os'):
            fig1=go.Figure()
            fig1.add_trace(go.Scatter(x=days,y=confcases,
                                    mode='lines+markers',
                                  marker_color = 'black',
                                  name='Scatter Plot'))
            fig1.add_trace(go.Bar(x=days,y=confcases,
                                  marker_color = 'orange',
                                  name='Bar Plot'))
            fig1.update_layout(title="Graph1",
                              paper_bgcolor=colors['background'],
                              plot_bgcolor = colors['background'],
                              font=dict(color='white'),
                              xaxis_title='No of Days',
                              yaxis_title='Confirmed people')

            fig2=go.Figure()
            fig2.add_trace(go.Pie(labels=category,values=ccases))
            fig2.update_layout(title="Graph2",
                              paper_bgcolor=colors['background'],
                              plot_bgcolor = colors['background'],
                              font=dict(color='white'))

            fig3=go.Figure()
            fig3.add_trace(go.Pie(labels=age,values=activecases,
                                 textinfo='label+percent',
                                 pull=[0,0.1,0,0]))
            fig3.update_layout(title="Graph3",
                              paper_bgcolor=colors['background'],
                              plot_bgcolor = colors['background'],
                              font=dict(color='white'))

            fig4=go.Figure()
            fig4.add_trace(go.Bar(x=gender,y=gcases,
                                 marker_color=['red','green']))
            fig4.update_layout(title="Graph8",
                              paper_bgcolor=colors['background'],
                              plot_bgcolor = colors['background'],
                              font=dict(color='white'),
                              xaxis_title='Male Vs Female',
                              yaxis_title='Raising cases among them')
            

    if(dropdown1=='bds'):
            fig1=go.Figure()
            fig1.add_trace(go.Scatter(x=day,y=cases,
                                 mode='lines+markers',
                                  marker_color = 'brown'))
            fig1.update_layout(title="Graph5",
                              paper_bgcolor=colors['background'],
                              plot_bgcolor = colors['background'],
                              font=dict(color='white'),
                              xaxis_title='No of Days',
                              yaxis_title='Confirmed people')

            fig2=go.Figure()
            fig2.add_trace(go.Pie(labels=cat,values=case))
            fig2.update_layout(title="Graph6",
                              paper_bgcolor=colors['background'],
                              plot_bgcolor = colors['background'],
                              font=dict(color='white'))

            fig3=go.Figure()
            fig3.add_trace(go.Pie(labels=age,values=activecases,
                                 textinfo='label+percent',
                                 pull=[0,0.1,0,0]))
            fig3.update_layout(title="Graph7",
                              paper_bgcolor=colors['background'],
                              plot_bgcolor = colors['background'],
                              font=dict(color='white'))

            fig4=go.Figure()
            fig4.add_trace(go.Bar(x=day,y=gcases,
                                 marker_color=['purple','magenta']))
            fig4.update_layout(title="Graph8",
                              paper_bgcolor=colors['background'],
                              plot_bgcolor = colors['background'],
                              font=dict(color='white'),
                              xaxis_title='Male Vs Female',
                              yaxis_title='Raising cases among them')

    return [fig1,fig2,fig3,fig4]
     
if __name__ == '__main__':
    app.run_server()
```
